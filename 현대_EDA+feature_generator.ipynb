```python
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.metrics import f1_score, classification_report
from sklearn.model_selection import StratifiedKFold as SKF
from imblearn.over_sampling import SMOTE, ADASYN
from xgboost import XGBRFClassifier, XGBClassifier

def macro_f1(y_true,y_pred):
    yt = y_true
    yp = y_pred.round()
#     print(yt, yp)
    return 1 - f1_score(
        yt, yp, 
        average = 'macro'
    )
```

```python
rf_params = {
    "n_estimators" : 1000,
    "n_jobs" : 24,
    "verbose" : 1,
    "criterion" : "entropy",
    "random_state" : None,
    "min_samples_split" : 8,
    "min_weight_fraction_leaf" : 0.0,
    "max_features" : "sqrt",
    "bootstrap" : True,
    "oob_score" : True,
    "class_weight" : {0:0.1, 1: 0.9}
}
```

```python

def data_preprocessing_fill0(df):
  df = pd.read_csv('./train.csv')
  train_df=df.drop(columns='ID',axis=1)
  train_df['COMPONENT_ARBITRARY']=train_df['COMPONENT_ARBITRARY'].astype('category').cat.codes
  test_df=pd.read_csv('./test.csv')
  test_df=test_df.drop(columns='ID')
  test_index=test_df.columns
  train_df=train_df.fillna(0)[test_index]

  return train_df

def data_preprocessing_fillmean(df):
  df = pd.read_csv('./train.csv')
  train_df=df.drop(columns='ID',axis=1)
  train_df['COMPONENT_ARBITRARY']=train_df['COMPONENT_ARBITRARY'].astype('category').cat.codes
  test_df=pd.read_csv('./test.csv')
  test_df=test_df.drop(columns='ID')
  test_index=test_df.columns
  train_df=train_df.fillna(df.mean())[test_index]

  return train_df

def data_preprocessing_ffill(df):
  df = pd.read_csv('./train.csv')
  train_df=df.drop(columns='ID',axis=1)
  train_df['COMPONENT_ARBITRARY']=train_df['COMPONENT_ARBITRARY'].astype('category').cat.codes
  test_df=pd.read_csv('./test.csv')
  test_df=test_df.drop(columns='ID')
  test_index=test_df.columns
  train_df=train_df.fillna(method= 'ffill')[test_index]

  return train_df

def data_preprocessing_bfill(df):
  df = pd.read_csv('./train.csv')
  train_df=df.drop(columns='ID',axis=1)
  train_df['COMPONENT_ARBITRARY']=train_df['COMPONENT_ARBITRARY'].astype('category').cat.codes
  test_df=pd.read_csv('./test.csv')
  test_df=test_df.drop(columns='ID')
  test_index=test_df.columns
  train_df=train_df.fillna(method= 'bfill')[test_index]

  return train_df
```

```python
train_df=data_preprocessing_fillmean(df)
```

```python
import os
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import TomekLinks

skf = SKF(shuffle=True)
save_dir = os.path.join('ckpt','v01')
qb_rfs = [RF(**rf_params) for _ in range(5)]
ovsample = ADASYN(n_jobs = 24)


for n,(train_idx, valid_idx) in enumerate(skf.split(train_df.values, df['Y_LABEL'].values)):
    print(n)
    train_x, train_y = train_df.values[train_idx], df['Y_LABEL'].values[train_idx]
    valid_x, valid_y = train_df.values[valid_idx], df['Y_LABEL'].values[valid_idx]

    rf = qb_rfs[n]
    rf.fit(*ovsample.fit_resample(train_x,train_y))

    y_hat = rf.predict(valid_x)
    print(classification_report(y_hat,valid_y))
#     print(f1_score(valid_y,y_hat, pos_label = 0))
#     print(f1_score(valid_y,y_hat, pos_label = 1))
#     print(f1_score(valid_y,y_hat, average = 'macro'))
```
